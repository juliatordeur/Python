Exercício 2


> Usando a tabela dados profissionais, escreva uma query que retorne todos os campos da tabela ordenando-os pela coluna tempo de casa, de maneira decrescente.



```python
cursor.execute("SELECT Id, graduacao, cargo, salario, tempo_casa FROM dados_profissionais ORDER BY tempo_casa DESC")
cursor.fetchall()
```

Exercício 4


> Usando a tabela dados cadastrais, escreva uma query que liste apenas os funcionários que possuem sobrenome "silva", ordenados pelo nome ocmpleto, ou seja, nome e sobrenome.



```python
cursor.execute("SELECT * FROM dados_cadastrais WHERE sobrenome = 'silva' ORDER BY nome ASC")
cursor.fetchall()
```

Exercício 5


> Escreva uma query que liste o valor da folha de pagamento dessa empresa por cargo, usando a tabela dados_profissionais (Ou seja, somar os salários em cada cargo).



```python
cursor.execute("SELECT SUM(salario) FROM dados_profissionais GROUP BY cargo ORDER BY salario ASC")
cursor.fetchall()
```
